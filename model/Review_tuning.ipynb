## Check the result of tuning

```python
import os
data_folder = os.path.join(os.getcwd(), 'data')
exp_name = '20220608_trial0'
ray_folder = os.path.join(data_folder,'ray_results')
exp_folder = os.path.join(ray_folder, exp_name)
```

```python
from ray.tune import ExperimentAnalysis
analysis = ExperimentAnalysis(exp_folder)
analysis.get_best_trial(metric="score",mode="min",).last_result
```

```python
import pdb
import sys
```

```python
paramset = []
k = 0
for i in range(len(analysis.trials)):
    try:
        if analysis.trials[i].last_result['score']!=2:
            u = {'score':analysis.trials[i].last_result['score'], 'ntau':analysis.trials[i].last_result['config']['ntau'], 'nsig':analysis.trials[i].last_result['config']['nsig'],
                 "c_ee":analysis.trials[i].last_result['config']['c_ee'], "c_ei":analysis.trials[i].last_result['config']['c_ei'],'c_ie':analysis.trials[i].last_result['config']['c_ie'],
                 'c_ii':analysis.trials[i].last_result['config']['c_ii'],'tau_e':analysis.trials[i].last_result['config']['tau_e'],'tau_i':analysis.trials[i].last_result['config']['tau_i'],
                 'r_e':analysis.trials[i].last_result['config']['r_e'],'r_i':analysis.trials[i].last_result['config']['r_i']}
            paramset.append(u)
            k+=1
    except Exception:
        continue
```

```python
import scipy
Fs=2**-1
region = config_one_region()
surface = config_surface(region)
sim = config_simulator(params, region, surface, integ_mode='deterministic', simu_length=210)
savg_time, savg_data = run_simulation(sim, params, return_signal=True)
savg_t = np.squeeze(np.array(savg_time))
savg = np.squeeze(np.array(savg_data))

# Check number of peaks to determin whether it's periodic or monotonic
waive_length = int(12/Fs) # Leaving out the first 5ms to get rid of artifact from the random seed
x_peak = scipy.signal.find_peaks(savg[waive_length:])
evaluation.plot_result(savg_t[waive_length+85:200],savg[waive_length+85:200], peaks=x_peak[0]-85)
```
